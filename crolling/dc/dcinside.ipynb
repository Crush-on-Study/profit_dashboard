```python
def find_low_page():
    div_a_tag = [i for i in range(3, 17)] ## paging_box a tag 인덱스
    select_page_list = []
    for a_idx in div_a_tag:
        paging_box = driver.find_elements(By.XPATH, f'//*[@id="container"]/section[1]/article[2]/div[4]/div[1]/a[{a_idx}]')
        paging_box[0].text
        select_page_list.append(paging_box[0].text)
    return select_page_list
```
